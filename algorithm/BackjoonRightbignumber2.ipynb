```python
class Stack:
    def __init__(self):
        self.items = []
        self.NGE = [-1]
        self.stack = []
    
    def reset(self,item):
        self.items = item  
    
    def findrightbig(self,N):
        self.stack.append(self.items.pop())

        for i in range(N-2,-1,-1):
            while self.stack:
                if self.items[i] < self.stack[-1]:
                    self.NGE.append(self.stack[-1])
                    self.stack.append(self.items[i])
                    break
                else:
                    self.stack.pop()
            if len(self.stack) == 0:
                self.stack.append(self.items[i])
                self.NGE.append(-1)
        
        self.NGE.reverse() 
        return  print(*self.NGE)
    
N = int(input())
item = list(map(int,input().split()))

s = Stack()
s.reset(item)
s.findrightbig(N)
```

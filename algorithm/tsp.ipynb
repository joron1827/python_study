```python
import numpy as np
import random
import csv



class TspProblem():
    def __init__(self):
        self.cities = []
        self.population = []
        self.total_cost = []
        self.givenSolution = []
        self.popSize = 10
        
    def distance(self,x, y):
        dist = np.linalg.norm(np.array(x)-np.array(y))
        return dist
        
    def readTSP(self,file):
        with open(file, mode='r', newline='') as tsp:
            reader = csv.reader(tsp)
            for i in reader:
                self.cities.append(i)
                
    def readSol(self,file):
        with open(file, mode='r', newline='') as solution:
            reader = csv.reader(solution)
            for i in reader:
                self.givenSolution.append(int(i[0]))
               
    def generatePopulation(self):
        initPop = [] # 모집단 초기화
        for i in range(self.popSize):
            random.shuffle(self.givenSolution)
            indexOfStart = self.givenSolution.index(0)
            front = self.givenSolution[indexOfStart:]
            back = self.givenSolution[:indexOfStart]
            initPop = front + back
            initPop.append(int(0))
            self.population.append(initPop)

    def fitnessEval(self):
        fitnessEval = []
        for i in range(self.popSize):
            total_cost = 0
            for idx in range(len(self.givenSolution)):
                pos_city_1 = [float(self.cities[self.population[i][idx]][0]), float(self.cities[self.population[i][idx]][1])]
                pos_city_2 = [float(self.cities[self.population[i][idx+1]][0]), float(self.cities[self.population[i][idx+1]][1])]
                dist = self.distance(pos_city_1, pos_city_2)
                
                total_cost += dist
            fitnessEval.append(total_cost)
            

        return fitnessEval
    
    def select(self,fitnessEval):
        c = np.cumsum(fitnessEval)
        parents = []
        for i in range(2):
            r = sum(fitnessEval) * np.random.rand()
            pick = np.argwhere(r <= c)
            parents.append(pick[0][0])
        return parents
    
    # def cyclecrossover(self,parents):
    #     child = []
    #     mom = self.population[parents[0]]
    #     dad = self.population[parents[1]]
        
        
        
        
        
   


a = TspProblem()
a.readTSP('TSP.csv')
a.readSol('example_solution.csv')
a.generatePopulation()
print(a.cities[0][1])
totalcost = a.fitnessEval()
print(totalcost)
a.select(totalcost)




    
        
    
```
